# Converting files

The script **converter.py** handles convertion from and to the following list of formats:
  * MED
  * SERAFIN
  * VTK
  * CGNS
  
To be able to do the conversion Telemac-Mascaret must be compiled wiht the optional library associated with the format.

By default only SERAFIN and VTK are handled.

```python
!converter.py -h
```

All the conversion follow the same structure.

Below is an example of the conversion of a selafin file and its boundary file into MED.

This will generate a MED file and a MED boundary file.

```python
HOMETEL = %env HOMETEL
%cd -q {HOMETEL}/examples/python3/converter

!converter.py srf2med mesh.slf -b mesh.cli mesh2.med
```
