## Extract bottom state from ptravers result file

The script `converter.py extract_ptravers_res_to_geoc` allows to extract a bottom and layer state from ptravers file to a Courlis geometry file (`.geoC`) which can be used to do a new Courlis simulation beginning from this state. It can be used also to perform a Mascaret simulation with a geometry from Courlis.


```python
!converter.py extract_ptravers_res_to_geoc -h
```

```python
# Run Soni
HOMETEL = %env HOMETEL
%cd -q {HOMETEL}/examples/courlis/Soni
!mascaret.py sarap.xcas
```

Output file (`-o` or `--output` arguments) is optional (an automatic name will be attributed `inputfilename_listingcourlis.opt`)

Option record alows to select the time frame to extract (by default last time step of ptravers file is extracted)

```python
!converter.py extract_ptravers_res_to_geoc resu.ptravers -r 50
```

To get the number of a time frame in a ptravers file, `MascaretFile.summary()` can be used (last section of the summary):

```python
import os.path
from data_manip.formats.mascaret_file import MascaretFile

courlis_file = MascaretFile(os.path.join(os.environ['HOMETEL'], 'examples', 'courlis', 'Soni', 'resu.ptravers'))
print(courlis_file.summary())
```
